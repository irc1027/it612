```python
%%javascript

function createBurner(secret, limit) {
    let count = 0;
    return function() {
        if (count < limit){
            count++;
            return secret;
        } else {
            return 'Message destroyed'
        }
    };
}

const msg = createBurner('fart', 2);
msg();
msg();
msg();
```
