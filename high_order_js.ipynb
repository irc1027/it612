```python
%%writefile code.js
// Part A
const data = [5, 12, 3, 20, 11, 9, 30];
const ndata = [];
for (let i = 0; i < data.length; i++) {
  if (data[i] > 10){
    ndata.push(data[i]*2);
  }
}
console.log(ndata);

const data1 = [5, 12, 3, 20, 11, 9, 30];
const ndata1 = data1.filter(item => item > 10).map(item => item * 2);
console.log(ndata1);


// Part B
const numbers = [1, 2, 3, 4, 5, 6];
const r = numbers.filter(n => n % 2 === 0).map(n => n * 2).reduce((sum, n) => sum + n, 0);
console.log(r);

// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];
console.log(users.some(u => u.age < 18));
console.log(users.find(u => u.id === 3));
users.forEach(u => console.log(u.name));

// Part D
function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);
  }
}
repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
